```python
#Cek apakah ada penurunan nilai 
import itertools
import operator

def monotone_increasing(lst):
    pairs = zip(lst, lst[1:])
    return all(itertools.starmap(operator.le, pairs))

def monotone_decreasing(lst):
    pairs = zip(lst, lst[1:])
    return all(itertools.starmap(operator.ge, pairs))

def monotone(lst):
    string_hasil = ""
    if monotone_increasing(lst):
      string_hasil =+ "Monoton Naik"
    elif monotone_decreasing(lst):
      string_hasil += "Monoton Turun"
    return string_hasil
```
